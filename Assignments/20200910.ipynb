### Problem statement

#### Functions to caluculate the following
- Numbers of lines in the file
- number of words in the file
- number of characters in the file
- The size of file

```python
def linecount(filename):
    lc=0
    f=open(filename,'r')
    for line in f:
        lc+=1
    f.close()
    return lc

def wordcount(filename):
    wc=0
    f=open(filename,'r')
    for line in f:
        words = line.split()
        wc += len(words)
        
    f.close()
    return wc

def charcount(filename):
    cc=0
    f=open(filename,'r')
    for line in f:
        for words in line:
            for letters in words:
                if(letters!="" and letters!='\n'):
                    cc+=1
    f.close()
    return cc

import os
def filesize(filename):
    file_size = os.path.getsize(filename)
    return file_size

filename='Datasets/sampletext'
lc=linecount(filename)
wc=wordcount(filename)
cc=charcount(filename)
fs=filesize(filename)
print(lc,wc,fs,cc)
```

*  Generate a Marks report



```python
def clsavg(filename):
    s=0
    c=0
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            m=int(i)
            s=s+m
            c+=1
    return s/c
def passpercentage(filename):
    c=0
    tc=0
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            tc+=1
            m=int(i)
            if(m>40):
                c+=1
    ppercent=(c/tc)*100
    return ppercent
def failpercentage(filename):
    c=0
    tc=0
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            tc+=1
            m=int(i)
            if(m<40):
                c+=1
    fpercent=(c/tc)*100
    return fpercent
def distinctpercentage(filename):
    c=0
    tc=0
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            tc+=1
            m=int(i)
            if(m>80):
                c+=1
    dpercent=(c/tc)*100
    return dpercent

def highestfrequency(filename):
    c=0
    m=[]
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            m.append(int(i))
    large=max(m)
    for j in m:
        if(large==j):
            c+=1
    return large,c         

def lowestfrequency(filename):
    c=0
    m=[]
    f=open(filename,'r')
    for line in f:
        mar=line.split(',')
        for i in mar:
            m.append(int(i))
    small=min(m)
    for j in m:
        if(small==j):
            c+=1
    return small,c

    
def marksreport(filename):
    f=open(filename,'w')
    filename1="Datasets/samplemarks.txt"
    avg=clsavg(filename1)
    pp= passpercentage(filename1)
    fp= failpercentage(filename1)
    dp=distinctpercentage(filename1)
    hf=highestfrequency(filename1)
    lf=lowestfrequency(filename1)
    l1="----MARKS REPORT-----"
    l2="Class Average : "
    l3="pass percentage :"
    l4="Fail percentage :"
    l5="Distinct percentage :"
    l6="Highest frequency :"
    l7="lowest mark frequency :"
    f.write('{}\n{}{}\n{}{}\n{}{}\n{}{}\n{}{}\n{}{}\n'.format(l1,l2,avg,l3,pp,l4,fp,l5,dp,l6,hf,l7,lf))
    f.close()
    
filename="Datasets/marksreport.txt"
marksreport(filename)
    
          
```
